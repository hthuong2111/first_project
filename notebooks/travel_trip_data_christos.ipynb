## Cleaning dataframe

### Starting with "destination"

```python
# I want to take the column destination, and rather split it in two, destination_city and destination_country

# Split on the first comma into two columns
travel_trip_christos_df[["destination_city", "destination_country"]] = (
    travel_trip_christos_df["destination"]
        .str.split(",", n=1, expand=True)
)

# Clean up whitespace
travel_trip_christos_df["destination_city"] = (
    travel_trip_christos_df["destination_city"].str.strip()
)
travel_trip_christos_df["destination_country"] = (
    travel_trip_christos_df["destination_country"].str.strip()
)

travel_trip_christos_df
```

```python
# I will drop the column "destination" after merging my table with the cleaned tables with my colleagues working in parallel.
```

```python
# will however first move the two new columns on the right and next to destination
```

```python
cols = list(travel_trip_christos_df.columns)
cols.insert(cols.index("destination") + 1, cols.pop(cols.index("destination_city")))
travel_trip_christos_df = travel_trip_christos_df.loc[:, cols]

cols = list(travel_trip_christos_df.columns)
cols.insert(cols.index("destination_city") - 1, cols.pop(cols.index("destination_country")))
travel_trip_christos_df = travel_trip_christos_df.loc[:, cols]

cols = list(travel_trip_christos_df.columns)
cols.insert(cols.index("destination_country") - 1, cols.pop(cols.index("destination")))
travel_trip_christos_df = travel_trip_christos_df.loc[:, cols]

travel_trip_christos_df
```

```python
# for transportation type and transportation_cost NaN values, we will group by destination_city and calculate the count of transportation_type and mean transportation_cost, and filter for Rome.

# we will need to cast the data type of the column transportation_cost to float, before we can proceed with grouping.

# for this to happen we will need to clean the column transportation_cost
```

```python
travel_trip_christos_df.transportation_cost.unique()
```

```python
# we will need to remove anything but the numbers
```

```python
travel_trip_christos_df.loc[:, "transportation_cost"] = (
    travel_trip_christos_df["transportation_cost"]
        .astype(str)                      # in case there are numbers/NaN in there
        .str.strip()
        .str.strip("$")
        .str.replace(",", "", regex=False)
        .str.replace(" USD", "", regex=False)
        .str.strip()
)

travel_trip_christos_df.loc[:, "transportation_cost"] = pd.to_numeric(
    travel_trip_christos_df["transportation_cost"],
    errors="coerce"
)

travel_trip_christos_df.transportation_cost.unique()
```

```python
# assuming travel_trip_christos_df was created with .copy() earlier
#travel_trip_christos_df.loc[:, "transportation_cost"] = (
#   travel_trip_christos_df["transportation_cost"]
#     .fillna("0")
# )

# travel_trip_christos_df.transportation_cost.unique()
```

```python
# now we cast, in order to be able to group later

travel_trip_christos_df.loc[:, "transportation_cost"] = (
    travel_trip_christos_df["transportation_cost"].astype("Int64")
)

travel_trip_christos_df.dtypes
```

```python
# for transportation type and transportation_cost NaN values, we will group by destination_city and calculate the count of transportation_type and mean transportation_cost, and filter for Rome.

grouped_df = (travel_trip_christos_df.groupby(["destination_city", "transportation_type"]).agg(transportation_type_count = ("transportation_type", "count"), transportation_cost_mean = ("transportation_cost", "mean")))
filtered_grouped_df = grouped_df.loc[["Rome"]]
filtered_grouped_df
```

```python
# therefore, we will fill the empty values transportation_type with "Train", and transportation_cost with "405.0"
travel_trip_christos_df.loc[:, "transportation_type"] = (
    travel_trip_christos_df["transportation_type"].fillna("Train")
)

travel_trip_christos_df.transportation_type.isnull().sum()

```

```python
# now let's replace the 0 we placed before in place of NaN, with the mean value.
travel_trip_christos_df.loc[:, "transportation_cost"] = (
    travel_trip_christos_df["transportation_cost"].fillna(405.0)
)

travel_trip_christos_df.transportation_cost.isnull().sum()
```

### Merging equivalent transportations types

```python
# ok, we will merge "Plane" and "Airplane" under "Airplane", and "Car rental" to "Rental Car", and "Car" to "Private Car"
travel_trip_christos_df.loc[:, "transportation_type"] = (
    travel_trip_christos_df["transportation_type"]
        .replace({"Plane": "Airplane"})
)

travel_trip_christos_df.loc[:, "transportation_type"] = (
    travel_trip_christos_df["transportation_type"]
        .replace({"Car rental": "Rental Car"})
)

travel_trip_christos_df.loc[:, "transportation_type"] = (
    travel_trip_christos_df["transportation_type"]
        .replace({"Car": "Private Car"})
)

travel_trip_christos_df["transportation_type"].unique()
```
